# Clueweb09-B Global Ordering Evaluation

```python
import pandas as pd

S = (pd.read_csv('/data/index/dai/full/cw09b.titles', names=['title'])
     .assign(rank=lambda df: list(reversed(range(len(df)))))
     .assign(docid=lambda df: list(range(len(df))))
     .assign(title=lambda df: df['title'].astype('category')))
```

```python
H = (pd.read_csv('/data/queries/clueweb/qrels.csv')
     .groupby('title')[['relevance']].max().reset_index()
     .rename(columns={'relevance': 'rank'})
     .astype({'title': S['title'].dtype}))
H = pd.merge(S[['title', 'docid']], H, how='left').fillna({'rank': 0}).sort_values('rank', ascending=False)
```

```python
dochits = pd.read_csv('/data/dochits/clueweb/500k/500k.dochits')
```

```python
titles = pd.read_csv('/data/dochits/clueweb/500k/cw09b.titles', names=['title'])
titles.index.name = 'docid'
titles.reset_index(inplace=True)
```

```python
dochits_titles = pd.merge(titles, dochits, on='docid').sort_values('h1000', ascending=False)
```

```python
dochits_titles.to_csv('/data/dochits/clueweb/500k/500k-title.dochits', index=False)
```

```python
(dochits_titles.sort_values('h100', ascending=False)['title']
 .to_csv('/data/dochits/clueweb/500k/500k.ordering-100', index=False, header=False))
```
