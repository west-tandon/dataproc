```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
import numpy as np
```

```python
feature_dir = '/data/michal/experiments/oss/cw09b-dai/features'
documents = pq.read_table(f'{feature_dir}/documents.parquet').to_pandas()
```

```python
documents
```

```python
results = pq.read_table(
    '/data/michal/experiments/oss/cw09b-dai/features/trec/cw09b-trec_eval-queries-OR.txt.results').to_pandas()
```

```python
shard_res = (pd.merge(results[['query', 'rank', 'gdocid']],
                     documents[['docid', 'shard']],
                     left_on='gdocid', right_on='docid')
            .drop(columns=['gdocid'])
            .sort_values(['query', 'rank']))
```

```python
shard_res.to_csv('/data/michal/experiments/loadbalancing/trec-results-with-shards.csv',
                 index=False)
```

```python
results_g = (pd.merge(results,
                     documents,
                     left_on='gdocid', right_on='docid')
             .sort_values(['query', 'rank']))
```
